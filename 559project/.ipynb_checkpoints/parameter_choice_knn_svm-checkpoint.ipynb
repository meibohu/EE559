```python
import sklearn
from sklearn.preprocessing import StandardScaler
import csv
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn import preprocessing
import data
from utils import Config
from math import pi
from math import exp
from math import sqrt
import pandas as pd
import seaborn as sn
from sklearn import metrics

from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.svm import SVC

from sklearn.neighbors import KNeighborsClassifier




def feature_selection(train_set, test_set):
    train_feature = [i[0:13] for i in train_set]
    train_label = [i[-1] for i in train_set]
    test_feature = [i[0:13] for i in test_set]
    test_label = [i[-1] for i in test_set]
    ##PCA
    #     pca = PCA(n_components=13)
    #     pca.fit(train_feature)
    #     train_feature = pca.transform(train_feature)
    #     test_feature = pca.transform(test_feature)
    ## selectkbest
    model = SelectKBest(k=8).fit(train_feature, train_label)
    train_feature = model.transform(train_feature)
    test_feature = model.transform(test_feature)

    return train_feature, train_label, test_feature, test_label


def SVC_model2(train_set, test_set, model):
    train_feature, train_label, test_feature, test_label = feature_selection(
        train_set, test_set)
    model.fit(train_feature, train_label)
    accuracy = model.score(test_feature, test_label)
    return accuracy

def KNN_modell2(train_set, test_set, model):
    train_feature, train_label, test_feature, test_label = feature_selection(
        train_set, test_set)
    # classifier = KNeighborsClassifier(n_neighbors=4)
    y_pred = model.fit(train_feature, train_label).predict(test_feature)
    #     gnb.fit(train_feature, train_label).
    accuracy = model.score(test_feature, test_label)
    return accuracy


#split into k
def cross_split(dataset, folds):
    dataset_split = list()
    dataset_copy = list(dataset)
    for i in range(len(folds)):
        fold = list()
        for index in folds[i]:
            fold.append(dataset_copy[index])
        dataset_split.append(fold)
    return dataset_split


#for SVM
def evaluate_algorithm2(dataset, model, folds, *args):
    scores = []
    datadset_split = cross_split(dataset, folds)
    for fold in datadset_split:
        train_set = list(datadset_split)
        test_set = fold
        train_set.remove(fold)
        train_set = sum(train_set, [])
        accuracy = SVC_model2(train_set, test_set, model)
        scores.append(accuracy)
    return scores


#for KNN
def evaluate_algorithm3(dataset, model, folds, *args):
    scores = []
    datadset_split = cross_split(dataset, folds)
    for fold in datadset_split:
        train_set = list(datadset_split)
        test_set = fold
        train_set.remove(fold)
        train_set = sum(train_set, [])
        accuracy = KNN_modell2(train_set, test_set, model)
        scores.append(accuracy)
    return scores


# def test(train, test, best_one):
#     accuracy, y_pred, test_label = best_one(train, test)
#     return accuracy, y_pred, test_label

dataset, folds = data.getResult(Config['train'])

#choose best parameter for SVM
ACC1 = []
c_arr = [1, 500, 1, 500, 1]
gamma_arr = ['scale', 'scale', 'auto', 'auto', 1]
kernel_arr = ['linear', 'linear', 'rbf', 'rbf', 'rbf']
for k in range(5):
    para_C = c_arr[k]
    para_gamma = gamma_arr[k]
    para_kernel = kernel_arr[k]
    # acc = np.zeros((10, 10))  #store average accuracies p on the validation set
    # dev = np.zeros(
    #     (10, 10))  #store the estimated standard deviation of accuracies
    model2 = SVC(C=para_C, gamma=para_gamma, kernel=para_kernel)
    accuracy = evaluate_algorithm2(dataset, model2, folds)
    print(np.mean(accuracy))
    ACC1.append(np.mean(accuracy))

# graph
name_list = ['linear1', 'linear2', 'rbf1', 'rbf2', 'rbf3']
plt.bar(range(len(ACC1)), ACC1, color='rgb', tick_label=name_list)
plt.show()
k_SVM = np.argwhere(ACC1 == np.max(ACC1))
print(k_SVM)

#choose parameter for knn
ACC2 = []
K_arr = [2, 4, 6, 8, 10]
for k in range(5):
    K_para = K_arr[k]
    # acc = np.zeros((10, 10))  #store average accuracies p on the validation set
    # dev = np.zeros(
    #     (10, 10))  #store the estimated standard deviation of accuracies
    model2 = KNeighborsClassifier(n_neighbors=K_para)
    accuracy = evaluate_algorithm3(dataset, model2, folds)
    print(np.mean(accuracy))
    ACC2.append(np.mean(accuracy))
# graph
name_list = ['k=2', 'k=4', 'k=6', 'k=8', 'k=10']
plt.bar(range(len(ACC2)), ACC2, color='b', tick_label=name_list)
plt.show()
k_KNN = np.argwhere(ACC2 == np.max(ACC2))
print(k_KNN)
```
